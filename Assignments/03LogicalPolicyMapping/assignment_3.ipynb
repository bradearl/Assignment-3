## Section 3 — Classify Logical Constraint Types

### Policy 1: IF General property is inactive for 5 years, THEN Presumed abandoned; report to state.
*   **Constraint Type:** Activation
*   **Explanation:** This rule states that if a certain condition (property inactivity) is met, then a specific action (reporting to the state) *must* be taken. The `if_1` for `report_standard_dormancy` implies this action.

### Policy 2: IF Unpaid wages remain unclaimed for 1 year, THEN Dormancy trigger reached for escheatment.
*   **Constraint Type:** Activation
*   **Explanation:** Similar to Policy 1, the condition of unclaimed wages for a year *activates* the requirement to consider them for escheatment, directly influencing the `report_unclaimed_wages` decision.

### Policy 3: IF Property value is $100 or more, THEN Mail notice to owner 60+ days before reporting.
*   **Constraint Type:** Activation
*   **Explanation:** The value of the property *activates* the obligation to send a due diligence notice. This directly corresponds to the `send_due_diligence_notice` decision.

### Policy 4: IF Individual items are valued at less than $100, THEN May report in aggregate without owner details.
*   **Constraint Type:** Complementary
*   **Explanation:** This implies that if you choose to report small-value items, and they are under \$100, then you *must* or *may* report them in aggregate (`aggregate_report_small_items`). It doesn't strictly 'activate' one decision based on another variable, but rather describes a condition under which a specific reporting method is allowed/preferred. If you choose to report small items, and they are under \$100, then this policy allows/encourages the aggregate reporting method.
*   **Clarification:** The word "May" suggests this is not a strict activation but an option. If 'May' means 'must if you want to report them at all in this way', then it's an activation. If it means 'you have the option to do this instead of reporting individually', it leans towards complementary depending on other rules.

### Policy 5: IF Holder is a non-insurance entity, THEN Report and remittance due by October 31st.
*   **Constraint Type:** Activation
*   **Explanation:** The type of entity (`non-insurance`) *activates* the specific reporting deadline. This directly affects the `meet_october_31_deadline` decision.

### Policy 6: IF Digital assets are inactive for 5 years, THEN Report and deliver in native form.
*   **Constraint Type:** Activation
*   **Explanation:** The inactivity of digital assets *activates* the requirement to report and deliver them in a specific format, relating directly to the `report_digital_assets` decision.

```python
import numpy as np
import re

decision_variables = decisions_df['variable'].tolist()
policy_names = [f"Policy {i+1}" for i, _ in enumerate(if_then_rules)]

# Initialize an empty DataFrame for the policy stack
policy_stack_data = []

for decision_idx, decision_var in enumerate(decision_variables):
    row_data = {"Decision": decision_var}
    decision_description = decisions_df.loc[decisions_df['variable'] == decision_var, 'description'].iloc[0]

    # Extract keywords from decision variable and description
    # Filter out common stop words and short words
    decision_terms = set(re.findall(r'\b\w+\b', decision_var.lower()))
    decision_terms.update(set(re.findall(r'\b\w+\b', decision_description.lower())))

    # Remove generic terms that won't help in matching unless explicitly added back
    generic_noise_terms = {'decide', 'whether', 'to', 'a', 'an', 'the', 'of', 'and', 'or', 'for', 'period', 'do', 'not'}
    decision_terms = {term for term in decision_terms if term not in generic_noise_terms and len(term) > 2}

    # Add back specific action words if they are primary indicators
    if 'report' in decision_var:
        decision_terms.add('report')
    if 'send' in decision_var:
        decision_terms.add('send')
        decision_terms.add('notice') # specific to send_due_diligence_notice
    if 'meet' in decision_var:
        decision_terms.add('meet')
        decision_terms.add('deadline')
    if 'aggregate' in decision_var:
        decision_terms.add('aggregate')
    if 'items' in decision_var: # relevant for small_items
        decision_terms.add('items')
    if 'property' in decision_var or 'dormancy' in decision_var: # relevant for standard_dormancy
        decision_terms.add('property')
        decision_terms.add('dormancy')
    if 'wages' in decision_var:
        decision_terms.add('wages')
        decision_terms.add('payroll')
    if 'digital' in decision_var:
        decision_terms.add('digital')
        decision_terms.add('assets')

    for policy_idx, policy_rule in enumerate(if_then_rules):
        policy_name = policy_names[policy_idx]

        match_found = False
        policy_rule_lower = policy_rule.lower()

        # Check for direct keyword overlap using individual terms
        for term in decision_terms:
            if term in policy_rule_lower:
                match_found = True
                break

        # Add some specific hardcoded matches/overrides for robustness
        if decision_var == 'report_standard_dormancy' and ('dormancy' in policy_rule_lower or 'abandoned' in policy_rule_lower or 'property' in policy_rule_lower):
            match_found = True
        if decision_var == 'report_unclaimed_wages' and ('wages' in policy_rule_lower or 'payroll' in policy_rule_lower or 'unclaimed' in policy_rule_lower):
            match_found = True
        if decision_var == 'send_due_diligence_notice' and 'notice' in policy_rule_lower:
            match_found = True
        if decision_var == 'aggregate_report_small_items' and ('aggregate' in policy_rule_lower or 'small' in policy_rule_lower or 'items' in policy_rule_lower):
            match_found = True
        if decision_var == 'meet_october_31_deadline' and ('october 31st' in policy_rule_lower or 'deadline' in policy_rule_lower or 'remittance' in policy_rule_lower):
            match_found = True
        if decision_var == 'report_digital_assets' and 'digital assets' in policy_rule_lower:
            match_found = True

        if match_found:
            row_data[policy_name] = 'X'
        else:
            row_data[policy_name] = '' # Use empty string for better readability in output

    policy_stack_data.append(row_data)

policy_stack_df = pd.DataFrame(policy_stack_data)

# Set 'Decision' column as index for better readability
policy_stack_df = policy_stack_df.set_index('Decision')

display(policy_stack_df)
```
